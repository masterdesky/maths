```python
OUTDIR = './out/'
```

```python
def f(x):
    return 2/np.pi * np.arcsin(np.sin(np.pi/2 * x))

def g(x):
    return np.arcsin(f(x))
    
def h(x):
    return np.sqrt(1 - f(x)**2) + np.pi/2

def p(x):
    return -h(x)
```

```python
x = np.linspace(-10,10,300000)
fx = f(x)
gx = g(x)
hx = h(x)
px = p(x)
```

```python
fix, axes = plt.subplots(figsize=(20,10),
                         facecolor='seashell', subplot_kw={'facecolor' : 'seashell'})

axes.axis('off')

#axes.plot(x,fx)
gx_c = cm.Reds(0.8)
axes.plot(x,gx,
          color=gx_c, lw=5)
hx_c = cm.Reds(0.8)
axes.plot(x,hx,
          color=hx_c, lw=5)
px_c = cm.Reds(0.8)
axes.plot(x,px,
          color=px_c, lw=5)

axes.set_xlim(-8,8)
axes.set_ylim(-4,4)

plt.savefig(OUTDIR + 'hearts.png',
            format='png',
            dpi=200,
            bbox_inches='tight')

plt.show()
```
